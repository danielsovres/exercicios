O Verificador de Palíndromos   
O Problema: Crie uma função que receba uma frase ou palavra e determine se ela é um palíndromo (se lida da mesma forma de trás para frente).

Dica: Você vai precisar ignorar espaços, pontuação e letras maiúsculas/minúsculas.

Exemplo: "A base do teto desaba" deve retornar True.

```python
def palindromo():
    """
    Serve para verificar se uma palavra (ou frase) pode ser lida de
    trás pra frente, ou de frente pra trás.
    Exemplo: 
    Input: A base do teto desaba
    invertida = abasedotetodesaba 
    """
    word_normal = input('Digite uma palavra para verificar se é um palíndromo: ').lower().replace(' ', '' )
    word_invertida = ''

    for letra in word_normal:
        word_invertida = letra + word_invertida

    if word_normal == word_invertida:
        print('Wow, isso dá pra ler de traz pra frente ou de "frente pra trás"!')
    else:
        print('Isso não é um palíndromo')
```

```python
palindromo()
```

```python
def palindromo():
    """
    Mesma função, mas retornando True ou False
    """
    word_normal = input('Digite uma palavra para verificar se é um palíndromo: ').lower().replace(' ', '' )
    word_invertida = ''

    for letra in word_normal:
        word_invertida = letra + word_invertida

    if word_normal == word_invertida:
        return True
    else:
        return False
```
