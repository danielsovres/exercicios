Faça um sistema que calcule um empréstimo imobiliário
Variáveis: Valor da casa, prazo (em anos), prestação, renda mensal (salário), capacidade de pagamento (30% do salário)

Opções:
Se a capacidade de pagamento for:
    Menos de 30% do Salário = Aprovado
    Mais de 30% do Salário = Rejeitado
    

```python
valor = float(input('Digite o valor do imóvel pretendido: '))
anos = int(input('Digite em quantos anos você quer pagar: '))
prestacao = valor / (anos*12)
salario = float(input('Digite o valor da sua renda mensal: '))
capacidade_pag = salario*0.3

print(f'\nPara financiar este imóvel no valor de {valor:.2f} em {anos:.2f} anos, será necessário uma prestação de {prestacao:.2f}\n'
f'Foi calculado que a sua capacidade atual de pagamento mensal é de {capacidade_pag:.2f}\n')

if capacidade_pag > prestacao:
    print(f'Seu empréstimo de {valor:.2f} foi APROVADO!')
else:
    print(f'Infelizmente, seu empréstimo de {valor:.2f} foi REJEITADO!')

```
