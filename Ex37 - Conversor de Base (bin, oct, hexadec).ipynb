escreva um programa que leia um número inteiro qualquer e peça para o usuário escolher qual será a base de conversão.

```python
num = int(input('Digite o número que você quer converter: '))
print('''Escolha uma das bases para conversão:
[ 1 ] Para converter para BINÁRIO
[ 2 ] Para converter para OCTAL
[ 3 ] Para converter para HEXADECIMAL''')
while True:
    opcao = int(input('Qual sua opção? ')) 
    if opcao in [1,2,3]:
        break
    else:
        print('\nOpção inválida! Tente novamente!')

if opcao == 1:
    print(f'O número {num} em BINÁRIO é {bin(num)[2:]}')
elif opcao == 2:
    print(f'O número {num} em OCTAL é {oct(num)[2:]}')
elif opcao == 3:
    print(f'O número {num} em HEXADECIMAL é {hex(num)[2:].upper()}')
```
