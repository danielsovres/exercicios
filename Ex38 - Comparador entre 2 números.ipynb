Objetivo:
Criar um programa simples que compare dois números (num1 e num2)
O programa deve definir qual número é maior

Se os números inteiros forem iguais, o sistema deve indicar a igualdade entre eles

```
num1 = input('Digite o primeiro número: ')
num2 = input('Digite o segundo número: ')

if num1>num2:
    print('O PRIMEIRO valor é maior!')
elif num1==num2:
    print('Os dois valores são IGUAIS.')
else:
    print('O SEGUNDO valor é maior!')
```
