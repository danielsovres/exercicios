Você tem essa lista de vendas:
vendas = [("maçã", 5), ("banana", 10), ("maçã", 3)]

O objetivo: Criar um dicionário que some tudo. No final, ele deve mostrar que você tem 8 maçãs e 10 bananas.

Dica:

Python
estoque = {}
# Use um for para percorrer a lista de vendas
for produto, quantidade in vendas:
    # Se o produto já está no estoque, soma. Se não, cria.
Consegue montar a lógica desse for?

```python
#forma 1) usando IN
vendas = [("maçã", 5), ("banana", 10), ("maçã", 3)]
estoque = {}

for produto, quantidade in vendas: #para cada produto e quantidade em vendas 
    if produto in estoque: # 1) Se o produto já estiver no estoque
        estoque[produto] += quantidade #pegue a quantidade do produto que já existe e some com a nova
    else: #2) se o produto não estiver no estoque
        estoque[produto] = quantidade #pegue o produto do vendas e jogue no estoque = que ele receba a quantidade inicial

print(f'Seu estoque é de {estoque}') #printar estoque final
```

```python
#forma 2) usando NOT IN
vendas = [("maçã", 5), ("banana", 10), ("maçã", 3)]
estoque = {}

for produto, quantidade in vendas: #para cada produto e quantidade em vendas 
    if produto not in estoque: # 1) Se o produto não estiver no estoque (nenhum estará, a princípio)
        estoque[produto] = 0 #coloque o produto do vendas e atribua o valor 0
    estoque[produto] += quantidade #2) se ele já existir no estoque (depois da criação), adicione a quantidade

print(f'Seu estoque é de {estoque}') #printar estoque final
```

```python
#forma 2) + Estoque iterativo com for
vendas = [("maçã", 5), ("banana", 10), ("maçã", 3)]
estoque = {}

for produto, quantidade in vendas: #para cada produto e quantidade em vendas 
    if produto not in estoque: # 1) Se o produto não estiver no estoque (nenhum estará, a princípio)
        estoque[produto] = 0 #coloque o produto do vendas e atribua o valor 0
    estoque[produto] += quantidade #2) se ele já existir no estoque (depois da criação), adicione a quantidade
        
for produto, quantidade in estoque.items(): #printar cada item separadamente
    print(f'Produto: {produto}: {quantidade}')
```

Criar uma função que conte quantas letras o usuário digitou, desprezando os espaços:

```python
def letras_digitadas_sem_espacos():
    entrada = input('Digite um texto para contar quantas letras você digitou: ').lower().replace(' ', '')
    contagem = len(entrada)
    print(f'O texto que você digitou contém {contagem} letras.')
```
